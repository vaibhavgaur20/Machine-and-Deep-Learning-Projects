# Q1

```python
class Solution:
    def merge(self, nums1: list[int], m:int, nums2: list[int], n:int) -> None:
        
        p1 = m-1
        p2 = n-1
        
        for p in range(n+m-1,-1,-1):
            
            if p2 <0:
                break
            if p1 >=0 and nums1[p1] > nums2[p2]:
                nums1[p] = nums1[p1]
                p1-=1
            else:
                nums1[p] = nums2[p2]
                p2-=1
        
        print(nums1)
        
        
        
```

```python
obj = Solution()
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3
obj.merge(nums1,m,nums2,n)
```

# Q2




#### NOTE :- In computer science, an in-place algorithm is an algorithm that operates directly on the input data struture without requiring extra space propotional to the input size. In other words, it modifies the input in-place, without creating a separate copy of the data structure.

```python
class Solution_rem_elem:
    def rem_elem(self, nums:list[int],val:[int])->int:
        
        writer =0;
        for reader in range(len(nums)):
            if (nums[reader] != val):
                nums[writer] = nums[reader]
                writer +=1
        return writer 
                
                
```

```python
nums = [0,1,2,2,3,0,4,2]
val = 2
obj_rem_elem = Solution_rem_elem()

k = obj_rem_elem.rem_elem(nums,val)

print(k)
```

# Q3

```python
class Solution_rem_duplic:
    def removeDuplicates(self,nums:list[int])->int:
        w = 1
        for i in range(1,len(nums)):
            if nums[i] != nums[i-1]:
                nums[w] = nums[i]
                w +=1
        return w
                                        
```

```python
nums = [0,0,1,1,1,2,2,3,3,4]

obj_rem_dup = Solution_rem_duplic()

k = obj_rem_dup.removeDuplicates(nums)
print(nums)
print(k)
```
