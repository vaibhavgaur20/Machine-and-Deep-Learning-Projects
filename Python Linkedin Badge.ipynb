# Q1. What is an abstract class?
## Ans: An abstract class exists only so that other "concrete" classes can inherit from the abstract class.

#### Reason to have a abstract class
- So that we can prevent user to make an object of an abstract class which in below example are Computer and Vehicle classes which are abstract classes and the user cannot create there objects.
- It ensures that concrete classes(i.e. normal class) **must** define the abstract methods present in Abstract classes they inherited from. 

- In simple words abstract classes are a list of must have methods. 

````python
# We are using Abstract Base Class(ABC) 
# beacuse by default python doesn't support abstract class
from abc import ABC, abstractmethod
````

# Q4. What statement about static methods is true?

## ANS: Static methods serve mostly as utility methods or helper methods, since they can't access or modify a class's state.

### class method

an instance(self) will only make changes to that object but if use classmethod it makes changes to entire class and doesn't matter how many objects that you create the changes will appear in all

````python
class Employee:
    
    num_of_emps =0 
    raise_amt = 1.04
    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.email = first + '.' + last + '@email.com'
        self.pay = pay
        
        Employee.num_of_emps += 1
        
    def fullname(self):
        return '{} {}'.format(self.first,self.last)
    
    def apply_raise(self):
        self.pay = int(self.pay * self.raise_amt)
    # instead of using instance variable self we use class(cls) as our first  argument
    # cls is the common convention
    @classmethod
    def set_raise_amt(cls,amount):
        cls.raise_amt = amount

emp_1 = Employee('Corey','Schafer', 50000)
emp_2 = Employee('Test','Employee',60000)

Employee.set_raise_amt(1.05)

print(Employee.raise_amt)
print(emp_1.raise_amt)
print(emp_2.raise_amt)
````

# Q15. Which of these is NOT a characteristic of namedtuples?

- Each member of a namedtuple object can be indexed to directly, just like in a regular tuple.
- namedtuples are just as memory efficient as regular tuples.

# Q21. What does the built-in map() function do?
## ANS:  It applies a function to each item in an iterable and returns the value of that function.


````python
from functools import reduce

li = [1,2,3,4,5,6]

filtered = list(filter(lambda n : n%2==0,li))

print(filtered)

ma_li = list(map(lambda n : n+2, filtered))

print(ma_li)

su = reduce(lambda a,b : a+b, ma_li)
print(su)
````
